```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline
import math
```

```python
df_SPARSE = pd.read_csv("combinedResults(SPARSE).csv")

df_dense = pd.read_csv("combinedResults.csv")

method1_df_SPARSE = df_SPARSE[['Vertices','Edges','Time Array (Method 1)']].copy()

method2_df_SPARSE = df_SPARSE[['Vertices','Edges','Time Array (Method 2)']].copy()

method1_df_dense = df_dense[['Vertices','Edges','Time Array (Method 1)']].copy()

method2_df_dense = df_dense[['Vertices','Edges','Time Array (Method 2)']].copy()
```

```python
# Plotting Method 1 and Method 2 for both sparse and dense graphs
plt.figure(figsize=(10, 6))

# Method 1 Sparse
plt.scatter(method1_df_SPARSE['Vertices'], method1_df_SPARSE['Time Array (Method 1)'], color='blue', label='Method 1 (Sparse)', alpha=0.6)
# Method 2 Sparse
plt.scatter(method2_df_SPARSE['Vertices'], method2_df_SPARSE['Time Array (Method 2)'], color='green', label='Method 2 (Sparse)', alpha=0.6)

# Method 1 Dense
plt.scatter(method1_df_dense['Vertices'], method1_df_dense['Time Array (Method 1)'], color='red', label='Method 1 (Dense)', alpha=0.6)
# Method 2 Dense
plt.scatter(method2_df_dense['Vertices'], method2_df_dense['Time Array (Method 2)'], color='orange', label='Method 2 (Dense)', alpha=0.6)

plt.title('Time Complexity of Methods 1 and 2 for Sparse and Dense Graphs')
plt.xlabel('Number of Vertices')
plt.ylabel('Time Taken (ms)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import math
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset
df_SPARSE = pd.read_csv('combinedResults(SPARSE).csv')

# Theoretical complexity functions
def matrix_array_complexity(V):
    return V ** 2  # Complexity for adjacency matrix with array-based priority queue

def list_heap_complexity(V, E):
    return (V + E) * math.log(V)  # Complexity for adjacency list with heap-based priority queue

# Applying the theoretical complexities to the dataframe
df_SPARSE['Matrix_Array_Complexity'] = df_SPARSE['Vertices'].apply(matrix_array_complexity)
df_SPARSE['List_Heap_Complexity'] = df_SPARSE.apply(lambda row: list_heap_complexity(row['Vertices'], row['Edges']), axis=1)

# Sort the dataframe by Vertices to make sure the line plots are correct
df_SPARSE = df_SPARSE.sort_values(by='Vertices')

# Plotting Method 1 (Empirical) vs Array (Theoretical) separately
plt.figure(figsize=(10, 6))

# Method 1 (Empirical) vs Array (Theoretical)
plt.scatter(df_SPARSE['Vertices'], df_SPARSE['Time Array (Method 1)'], color='blue', label='Method 1 (Empirical)', alpha=0.6)
plt.plot(df_SPARSE['Vertices'], df_SPARSE['Matrix_Array_Complexity'], color='red', linestyle='--', label='Theoretical Array (Sparse)', linewidth=2)

plt.title('Time Complexity: Method 1 (Empirical) vs Array (Theoretical)')
plt.xlabel('Number of Vertices')
plt.ylabel('Time Taken / Theoretical Complexity')
plt.legend()
plt.grid(True)
plt.show()

# Plotting Method 2 (Empirical) vs Heap (Theoretical) separately
plt.figure(figsize=(10, 6))

# Method 2 (Empirical) vs Heap (Theoretical)
plt.scatter(df_SPARSE['Vertices'], df_SPARSE['Time Array (Method 2)'], color='green', label='Method 2 (Empirical)', alpha=0.6)
plt.plot(df_SPARSE['Vertices'], df_SPARSE['List_Heap_Complexity'], color='orange', linestyle='--', label='Theoretical Heap (Sparse)', linewidth=2)

plt.title('Empirical vs Theoretical Time Complexity: Method 2 (Empirical) vs Heap (Theoretical)')
plt.xlabel('Number of Vertices')
plt.ylabel('Time Taken / Theoretical Complexity')
plt.legend()
plt.grid(True)
plt.show()

```
