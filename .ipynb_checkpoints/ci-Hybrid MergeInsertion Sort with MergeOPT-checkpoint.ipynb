```python
import random
import pandas as pd
```

```python
hybrid_df = pd.DataFrame(columns = ['n', 'number_of_key_comparisons'])
```

```python
def generate_random_array(length, min_value=0, max_value=1000000):
    """
    Generate a random array of integers.
    
    Parameters:
    length (int): Length of the array.
    min_value (int): Minimum value for array elements.
    max_value (int): Maximum value for array elements.

    Returns:
    list: A list of random integers.
    """
    return [random.randint(min_value, max_value) for _ in range(length)]

def check_if_sorted(arr):
    not_sorted = False
    for i in range(len(arr) - 1):
        if arr[i] > arr[i + 1]:
            not_sorted = True
            break
    if not_sorted:
        print("NOT SORTED")
    else:
        print("SORTED")
```

```python
def insertion_sort(a_list, start, end):
    global comparison_count
    for i in range(start + 1, end + 1):
        for j in range(i, start, -1):
            comparison_count += 1
            if a_list[j] < a_list[j - 1]:
                a_list[j], a_list[j - 1] = a_list[j - 1], a_list[j]
            else:
                break


def merge(slot, start, midpointer, end):
    global comparison_count
    n = start
    m = end
    mid = midpointer
    a = n
    b = mid + 1

    if m - n <= 0:
        return

    while a <= mid and b <= m:
        comparison_count += 1
        if slot[a] < slot[b]:  # No swap needed #ZW changed
            a += 1
        elif slot[a] > slot[b]:  # slot[a] > slot[b]
            tmp = slot[b]
            b += 1
            for i in range(mid + 1, a, -1):
                slot[i] = slot[i - 1]
            slot[a] = tmp
            a += 1
            mid += 1
        else:  # slot[a] == slot[b]
            if a == mid and b == m:
                break
            tmp = slot[b]
            b += 1
            a += 1
            for i in range(mid + 1, a, -1):
                slot[i] = slot[i - 1]
            slot[a] = tmp
            a += 1
            break


def mergeOPT(slot, start, midpointer, end):
    global comparison_count
    left = slot[start:midpointer+1]
    right = slot[midpointer+1:end+1]
    '''
    print(start)
    print(end)
    print(left)
    print(right)
    '''

    i = 0
    j = 0
    k = start

    while i < len(left) and j < len(right):
        comparison_count += 1
        if left[i] <= right[j]:
            slot[k] = left[i]
            k += 1
            i += 1
        else:
            slot[k] = right[j]
            k += 1
            j += 1

    while i < len(left):
        slot[k] = left[i]
        k += 1
        i += 1

    while j < len(right):
        slot[k] = right[j]
        k += 1
        j += 1

    '''
    print(slot)
    print()
    '''

def merge_insertion_sort(slot, start, end, threshold):
    if end - start <= threshold:
        insertion_sort(slot, start, end)
    else:
        mid = (start + end) // 2

        merge_insertion_sort(slot, start, mid, threshold)
        merge_insertion_sort(slot, mid + 1, end, threshold)

        mergeOPT(slot, start, mid, end)


def hybrid_merge_insertion_sort(arr, threshold=8):
    global comparison_count
    comparison_count = 0
    merge_insertion_sort(arr, 0, len(arr) - 1, threshold)
    return arr
```

```python
# Original Merge Sort Implementation
def merge_sort(arr, start, end):
    if start < end:
        mid = (start + end) // 2
        merge_sort(arr, start, mid)
        merge_sort(arr, mid + 1, end)
        mergeOPT(arr, start, mid, end)


def standard_merge_sort(arr):
    global comparison_count
    comparison_count = 0
    merge_sort(arr, 0, len(arr) - 1)
    return arr
```

```python
# Fixed S = 8, diffrent n, increase by 4000
n = 1000
j = 0

while n <= 10000000:
    print(n)
    ave_comparison_counts_hybrid = 0
    
    for i in range(3):
        arr = generate_random_array(n)

        # Reset and use hybrid_merge_insertion_sort
        global comparison_count
        comparison_count = 0
        sorted_arr_hybrid = hybrid_merge_insertion_sort(arr.copy())
        ave_comparison_counts_hybrid += comparison_count
    
    ave_comparison_counts_hybrid //= 3
    
    # Output the number of comparisons for each array and each method
    print("Hybrid Merge-Insertion Sort Comparisons:")
    print(f"Average: {ave_comparison_counts_hybrid}")
    hybrid_df.loc[j] = [n, ave_comparison_counts_hybrid]
    
    j += 1
    n += 9000

display(hybrid_df)
```
